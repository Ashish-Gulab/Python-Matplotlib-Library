**MATPLOTLIB**

Matplotlib pyplot is a plotting library used for 2D graphics in python programming language.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[4,5,1]
# (x1,y1)(1,4)(2,5)(3,1)
plt.plot(x,y,label='line one')
plt.title("This is a line")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend() # We can give the "line one" inside label
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') # "ro" r indicates red colour and o indicate round shape
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'r*') # "ro" r indicates red colour and * indicate star shape
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'r+', label='line1') # "ro" r indicates red colour and + indicate + shape
plt.plot([1,2,3,4],[1,2,3,4],'go-', label='line2') # '-' is used to join the points
plt.grid() # grid is used to show square lines
plt.legend() # defalut: it selects the best suitable place
plt.show()
```

```python
# Bar graph : compare the data among the different categories and longer the bar has the greater value
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,90])
plt.show()
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,90])
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60])
plt.show()
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,90],width=.3,label="BMW")
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],width=.4,label='Audi')
plt.xlabel('Days')
plt.ylabel('Distance')
plt.legend()
plt.show()
```

```python
# Scatter Graph : comapre variables
x=[1,1.5,2,2.5,3,4,5]
y=[7,8,8.5,10,10.5,12,13]
plt.scatter(x,y,label='income')
plt.legend()
plt.show()
```

```python
x=[1,1.5,2,2.5,3,4,5]
y=[7,8,8.5,10,10.5,12,13]
x1=[10,20,30,40,50]
y1=[2,3,4,5,6]
plt.scatter(x,y,label='income')
plt.scatter(x1,y1,color='r')
plt.show()
```

```python
# histogram graph : These are used to show a distribution. It is used to get the frequency of a particular data.
import numpy as np
student_age=[69,18,40,69,69,18,18,18,21,22,23,21,18,69,22,25,22,69,22,21,40,18,40,40,20,22,23,25]
bins=np.arange(1,100)
plt.hist(student_age,bins)
plt.show()
```

```python
# Pie Chart
days=[1,2,3,4,5]
slices=[2,2,3,13] # the values in slices is used to give the number of parts of the chart (as here 4 parts)
activities=['sleeping','eating','working','playing']
plt.pie(slices,labels=activities)
plt.show()
```

```python
# Working with multiple plots
#subplots
t1=np.arange(0.0,5,0.1)
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)
plt.plot(t1,f(t1))
plt.plot(t1,np.cos(2*np.pi*t1))
plt.show()
```
